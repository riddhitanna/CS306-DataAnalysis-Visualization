Plotting the normalized histograms for TMAX and TMIN for a particular station and comparing them for different bin sizes.

```python
#Importing all the required libraries
import pandas as pd 
import numpy as np 
import seaborn as sns 
import plotly.express as px
import matplotlib.pyplot as plt
import datetime
```

```python
#Reading the csv file 
df = pd.read_csv('/Users/riddhi/Downloads/Temperature_2020.csv') # or insert file path 
```

```python
station = input('Enter the name of the station for which you want the histograms:') #station for plotting
```

```python
df_c = df.copy()
df_copy = df_c.loc[df['STATION_NAME'] == station] #filtering the dataset to get values for just the 
#required station

#Removing invalid temperature value(s) - here, -9999

for i in df_copy.index:
    if(df_copy['TMAX'][i] ==-9999):
        df['TMAX'][i] = 0
for i in df_copy.index:
    if(df_copy['TMIN'][i] ==-9999):
        df_copy['TMIN'][i] = 0
```

```python
tmax_max = df_copy['TMAX'].max()
tmax_min = df_copy['TMAX'].min()
tmin_max = df_copy['TMIN'].max()
tmin_min = df_copy['TMIN'].min()
mean_tmax = df_copy['TMAX'].mean()
mean_tmin = df_copy['TMIN'].mean()
median_tmax = df_copy['TMAX'].median()
median_tmin = df_copy['TMIN'].median()
df_copy.describe()
```

### In the following cells, we plot the histogram for TMAX at the particular station for different number of bins.

```python
bin_width_tmax = 5 #bin width => Number of bins = (end-start)/bin width 
nbins = int((tmax_max-tmax_min)/bin_width_tmax)+1
fig, ax = plt.subplots(figsize=[17,8])
plt.style.use('ggplot')
fig = plt.hist(df_copy['TMAX'], bins = nbins, alpha=0.9, density=True, edgecolor='black')
plt.title('Normalized histogram with bin width = {} and number of bins = {} for TMAX at station {}'.format
          (bin_width_tmax,nbins,station), fontsize=20)
plt.grid(axis='y', alpha=1)
plt.xlabel('Temperature (in K)',fontsize=20)
plt.ylabel('Frequency (normalized)',fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axvline(mean_tmax, linestyle = '--', color = 'black', label='Mean = {}'.format(mean_tmax))
plt.axvline(median_tmax, linestyle = '--', color = 'blue', linewidth = 0.7,label='Median = {}'.format(median_tmax))
plt.legend(shadow=1, facecolor='white')
plt.show()
```

# Conclusion

From the above histograms, it is clear that the number of bins affects the interpretability of the data to a large extent. We can see that if we do not have the right number of bins, then the following two things can happen:

1. If the number of bins is higher than ideal (smaller bin width), then we cannot identify if there are any local incidences of high or low frequency and we cannot group the data properly to identify trends
2. If the number of bins is lower than ideal (larger bin width), we cannot see the distribution and variation of the data well. 

As we can see in the cases of both TMIN and TMAX, when we have taken higher number of bins, we cannot identify any groups or trends in the temperature. 
When we have taken lesser number of bins, we cannot see the variation of temperature well, and even though we can see where the mean and median falls, we do not know how exactly the temperature is distributed. 
Hence, when we take the ideal number of bins, we can group the data well and also identify trends and see how the temperatures are distributed. 

In the third histogram in both TMAX and TMIN, we can see that for TMAX, the data is a little right skew in the distribution and for TMIN, there is a little left skew in the distribution. This is as we would expect for maximum and minimum temperature distributions. We cannot observe this in the first two histograms in both the cases.
